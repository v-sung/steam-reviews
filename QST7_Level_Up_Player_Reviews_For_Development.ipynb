## a. Problem Statement

Steam’s massive volume of reviews and game data contains valuable insights into how players evaluate gameplay, value, and experience—but neither users nor developers can easily extract these patterns at scale. Players struggle to discover games that match their preferences, while developers lack clear, data-driven guidance on which features, genres, and pricing strategies drive positive sentiment and long-term success.

This project aims to bridge that gap by analyzing large-scale review data and game attributes to identify the factors that most influence player sentiment and engagement. By translating review patterns into actionable insights, we help developers understand what players expect, how to improve game quality, and how to position their products more effectively. At the same time, these insights support recommendation tools that help players find games aligned with their interests, improving decision-making across the Steam ecosystem.

## b. Motivation

As one of the world’s largest digital game distribution platforms, Steam hosts a massive user ecosystem and an incredibly diverse catalog of games. Yet the sheer volume of titles and the complexity of user behavior make it difficult for both players and developers to identify the insights that truly matter: What do players actually value? Which factors drive satisfaction or frustration? And how can developers optimize design, pricing, and content updates to better meet player expectations?

The motivation behind this notebook is to transform large-scale player reviews into a clear, data-driven, and actionable analytical framework. Steam reviews capture both sentiment and market performance, revealing patterns that help developers understand genre expectations, perceived value, engagement behavior, and how sentiment evolves over time. By analyzing these signals at scale, we provide insights that enable players, developers, marketers, and researchers to grasp key trends without needing to dig through millions of raw reviews themselves.

## c. Executive Summary

Our project analyzes over 51 million English Steam reviews combined with metadata from more than 111,000 games to understand how player sentiment, engagement, and game characteristics interact across the platform. This large-scale dataset provides a macro-level view of what players value, how they respond to different genres and pricing strategies, and how these patterns can guide developers in designing games that better meet audience expectations.

Across the platform, several consistent trends emerge. Player sentiment is broadly positive, making negative reviews a strong signal of deeper issues in gameplay, monetization, or overall execution. Genre plays a major role in shaping expectations: creative tools and utility-focused titles tend to receive consistently high satisfaction, while genres such as Action, Adventure, and RPG exhibit much wider variation, reflecting the importance of strong execution in meeting player standards. Pricing shows little relationship to sentiment—low-priced games perform just as well as premium-priced games—highlighting that perceived value, rather than absolute price, drives player satisfaction. Additionally, popularity and quality do not always align; high concurrent user counts do not strongly correlate with positive reviews, and many niche titles outperform mainstream games by delivering well-targeted experiences to specific audiences.

Sentiment also evolves over time, responding noticeably to updates, patches, and content releases. This finding underscores the value of continuous monitoring for live-service and early-access titles. Machine learning methods applied to the dataset illustrate how developers can use clustering and early engagement signals to forecast sentiment trends, identify potential quality issues, and benchmark their games against broader market behavior.

Taken together, these insights offer a data-driven framework that developers can use to align features with genre expectations, make more informed pricing and value-delivery decisions, prioritize updates that meaningfully influence sentiment, and incorporate review analytics into both pre-launch planning and post-launch support. By analyzing player feedback at scale, this project demonstrates how developers and users alike can benefit from a clearer understanding of the factors that shape game perception and long-term success on Steam.

Load the reviews dataset.

```python
path = "gs://qst843_team7_bucket/all_reviews.csv"
raw = (spark.read.option("header", True).option("escape",'"').option("nullValue", "NULL").csv(path))

reviews = (raw
    .withColumn("recommendationid", F.col("recommendationid").cast("long"))
    .withColumn("appid",            F.col("appid").cast("long"))
    .withColumn("author_steamid",   F.col("author_steamid").cast("long"))
    .withColumn("author_num_games_owned", F.col("author_num_games_owned").cast("int"))
    .withColumn("author_num_reviews",     F.col("author_num_reviews").cast("int"))
    .withColumn("author_playtime_forever",        F.col("author_playtime_forever").cast("long"))
    .withColumn("author_playtime_last_two_weeks", F.col("author_playtime_last_two_weeks").cast("long"))
    .withColumn("author_playtime_at_review",      F.col("author_playtime_at_review").cast("long"))
    .withColumn("votes_up",     F.col("votes_up").cast("long"))
    .withColumn("votes_funny",  F.col("votes_funny").cast("long"))
    .withColumn("weighted_vote_score", F.col("weighted_vote_score").cast("float"))
    .withColumn("comment_count", F.col("comment_count").cast("long"))
    .withColumn("timestamp_created",
                F.to_timestamp(F.from_unixtime(F.col("timestamp_created").cast("long"))))
    .withColumn("timestamp_updated",
                F.to_timestamp(F.from_unixtime(F.col("timestamp_updated").cast("long"))))
    .withColumn("author_last_played",
                F.to_timestamp(F.from_unixtime(F.col("author_last_played").cast("long"))))
    .withColumn("voted_up",
        F.when(F.lower(F.col("voted_up")).isin("true","1"), True)
         .when(F.lower(F.col("voted_up")).isin("false","0"), False)
         .otherwise(None).cast("boolean"))
    .withColumn("steam_purchase",
        F.when(F.lower(F.col("steam_purchase")).isin("true","1"), True)
         .when(F.lower(F.col("steam_purchase")).isin("false","0"), False)
         .otherwise(None).cast("boolean"))
    .withColumn("received_for_free",
        F.when(F.lower(F.col("received_for_free")).isin("true","1"), True)
         .when(F.lower(F.col("received_for_free")).isin("false","0"), False)
         .otherwise(None).cast("boolean"))
    .withColumn("written_during_early_access",
        F.when(F.lower(F.col("written_during_early_access")).isin("true","1"), True)
         .when(F.lower(F.col("written_during_early_access")).isin("false","0"), False)
         .otherwise(None).cast("boolean"))
    .withColumn("hidden_in_steam_china",
        F.when(F.lower(F.col("hidden_in_steam_china")).isin("true","1"), True)
         .when(F.lower(F.col("hidden_in_steam_china")).isin("false","0"), False)
         .otherwise(None).cast("boolean"))
)
```

```python
total_reviews = reviews_and_games.count()

posneg = (
    reviews_and_games
    .groupBy("voted_up")
    .count()
    .withColumn("share", F.col("count") / F.lit(total_reviews))
    .orderBy(F.desc("count"))
)

posneg.show(truncate=False)

posneg_readable = posneg.withColumn(
    "label",
    F.when(F.col("voted_up") == 1, F.lit("positive"))
     .otherwise(F.lit("negative"))
).select("label", "count", F.round("share", 4).alias("share"))
posneg_readable.show(truncate=False)
```

```python
from pyspark.sql import functions as F

per_game = (
    reviews_and_games
    .groupBy("appid", "name", "genres", "price")
    .agg(
        F.count(F.lit(1)).alias("review_count"),
        F.sum(F.col("voted_up").cast("int")).alias("positive_count")
    )
    .withColumn("positive_rate", F.col("positive_count") / F.col("review_count"))
    .orderBy(F.desc("review_count"))
)

per_game.cache()
per_game.show(10, truncate=False)

per_game.select("review_count").describe().show()

quantiles = per_game.approxQuantile("review_count", [0.5, 0.9, 0.95, 0.99], 0.01)
print({"median": quantiles[0], "p90": quantiles[1], "p95": quantiles[2], "p99": quantiles[3]})
```

```python
def top_n_games(df, n=10):
    return (
        df.orderBy(F.desc("review_count"))
          .limit(n)
          .select("appid", "name", "review_count", "positive_count", F.round("positive_rate", 3).alias("positive_rate"))
    )

for n in [10, 25, 50, 100]:
    print(f"
=== TOP {n} GAMES BY REVIEW COUNT ===")
    top_n_games(per_game, n).show(truncate=False)

top10_pd = top_n_games(per_game, 10).toPandas()

plt.figure(figsize=(10, 5))
sns.barplot(data=top10_pd, y="name", x="review_count", hue="name", palette="muted", legend=False)
plt.title("Top 10 Games by Review Count")
plt.xlabel("Number of Reviews")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

Counter-Strike: Global Offensive overwhelmingly dominates in review counts, indicating its massive player base and sustained popularity compared to other top games like Terraria and Team Fortress 2.

## **2b. Who are the top publishers by average review sentiment?**

```python
publisher_sentiment = (
    reviews_and_games
    .select("appid", "name", "publishers", "voted_up")
    .filter((F.col("voted_up") == False) | ((F.col("voted_up") == True))) # need this to clean data
)

publisher_sentiment = (
    publisher_sentiment
    .groupBy("publishers")
    .agg(
        sum(when(F.col("voted_up") == True, 1).otherwise(0)).alias("positive"),
        sum(when(F.col("voted_up") == False, 1).otherwise(0)).alias("negative"),
        F.count("*").alias("total_reviews")
    )
    .withColumn("percent_pos", (F.col("positive") / (F.col("positive") + F.col("negative"))))
    .filter(F.col("total_reviews") > 10000)
    .orderBy("percent_pos", ascending=False)
)
```

```python
publisher_sentiment_pd = publisher_sentiment.toPandas()

top_publishers = publisher_sentiment_pd.head(20)
```

```python
plt.figure(figsize=(10, 8))
ax = sns.catplot(data=top_publishers, kind="bar", x="percent_pos", y="publishers")
plt.title("Sentiment by Publisher")
plt.xlabel("Percent Positive")
plt.ylabel("Publisher")
plt.tight_layout()
plt.show()
```

## **2c. For a given game how do ratings/sentiment change over time?**

```python
from pyspark.sql.functions import year, quarter, when, col, sum as _sum, concat, lit

nov22_games = (
    games
    .select("appid", "name", "release_date")
    .filter(F.col("release_date") < "2022-11-30")
    .sample(False, 0.0005, 42) # adjust sample depending on size of dataset
)

nov22_games = (
    nov22_games
    .join(eng_reviews, "appid", "inner")
    .select("appid", "name", "timestamp_created", "voted_up")
    .filter((F.col("voted_up") == False) | ((F.col("voted_up") == True))) # need this to clean data
)

# code for months... better if reviews span a short range of time
"""nov22_games = (
    nov22_games
    .withColumn("timestamp_year", F.year("timestamp_created"))
    .withColumn("timestamp_month", F.month("timestamp_created"))
    .groupBy("appid", "name", "timestamp_year", "timestamp_month")
    .agg(
        sum(when(F.col("voted_up") == True, 1).otherwise(0)).alias("positive"),
        sum(when(F.col("voted_up") == False, 1).otherwise(0)).alias("negative"),
    )
    .withColumn("date", F.concat(col("timestamp_year"), F.lit("-"), F.col("timestamp_month")))
    .withColumn("percent_pos", (col("positive") / (col("positive") + col("negative"))))
    .orderBy("date")
)"""

# code for quarters... better if reviews span a long range of time
nov22_games = (
    nov22_games
    .withColumn("timestamp_year", F.year("timestamp_created"))
    .withColumn("timestamp_quarter", F.quarter("timestamp_created"))
    .groupBy("appid", "name", "timestamp_year", "timestamp_quarter")
    .agg(
        _sum(when(F.col("voted_up") == True, 1).otherwise(0)).alias("positive"),
        _sum(when(F.col("voted_up") == False, 1).otherwise(0)).alias("negative"),
    )
    .withColumn("date", F.concat(col("timestamp_year"), F.lit("-Q"), F.col("timestamp_quarter")))
    .withColumn("percent_pos", (col("positive") / (col("positive") + col("negative"))))
    .orderBy("date")
)
```

```python
nov22_games_pd = nov22_games.toPandas()
```

# **3. Machine Learning**

## 3a. K-Means Clustering - Segmenting games into clusters and examining characteristics of each cluster.

```python
from pyspark.ml import Pipeline
from pyspark.ml.feature import VectorAssembler, StandardScaler, PCA
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import ClusteringEvaluator
from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType
```

```python
reviews_and_games_k = reviews_and_games.withColumn(
    "voted_up_int",
    F.col("voted_up").cast("int")
)

games_for_kmeans = (
    reviews_and_games_k
        .groupBy("appid", "name", "genres", "categories", "price")
        .agg(
            F.avg("voted_up_int").alias("pct_positive_reviews"),
            F.count("*").alias("num_reviews"),
            F.avg("author_num_reviews").alias("avg_author_num_reviews"),
            F.avg("author_num_games_owned").alias("avg_author_games_owned")
        )
        # Filter out games with very few reviews
        .where(F.col("num_reviews") >= 20)
)

feature_cols = [
    "price",
    "pct_positive_reviews",
    "num_reviews",
    "avg_author_num_reviews",
    "avg_author_games_owned"
]

games_for_kmeans = games_for_kmeans.na.drop(subset=feature_cols)
games_for_kmeans.show(5, truncate=False)
games_for_kmeans.printSchema()
```

```python
assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features_assembled"
)

scaler = StandardScaler(
    inputCol="features_assembled",
    outputCol="features_scaled",
    withMean=True,
    withStd=True
)

pca = PCA(
    k=2,
    inputCol="features_scaled",
    outputCol="pca"
)

pipeline = Pipeline(stages=[assembler, scaler, pca])
pipeline_model = pipeline.fit(games_for_kmeans)
transformed = pipeline_model.transform(games_for_kmeans)
transformed.select("appid", "name", "features_assembled", "pca").show(5, truncate=False)
```

```python
kmeans = KMeans(
    k=3,
    seed=843,
    featuresCol="pca",
    predictionCol="cluster"
)

model = kmeans.fit(transformed)

predictions = model.transform(transformed)

predictions.select(
    "appid", "name", "pct_positive_reviews", "num_reviews", "price", "cluster"
).show(10, truncate=False)
```

```python
evaluator = ClusteringEvaluator(
    featuresCol="pca",
    predictionCol="cluster",
    metricName="silhouette",
    distanceMeasure="squaredEuclidean"
)

silhouette = evaluator.evaluate(predictions)
print("Silhouette with squared euclidean distance = " + str(silhouette))

centers = model.clusterCenters()
print("Cluster Centers:")
for idx, center in enumerate(centers):
    print(f"Cluster {idx}: {center}")
```

```python
cluster_summary = (
    predictions
      .groupBy("cluster")
      .agg(
          F.count("*").alias("num_games"),
          F.avg("pct_positive_reviews").alias("avg_pct_positive_reviews"),
          F.avg("num_reviews").alias("avg_num_reviews"),
          F.avg("price").alias("avg_price"),
          F.avg("avg_author_num_reviews").alias("avg_author_num_reviews"),
          F.avg("avg_author_games_owned").alias("avg_author_games_owned")
      )
      .orderBy("cluster")
)

cluster_summary.show(truncate=False)
```

```python
firstElement = udf(lambda v: float(v[0]), FloatType())
secondElement = udf(lambda v: float(v[1]), FloatType())

plot_df = (
    predictions
        .withColumn("pc1", firstElement("pca"))
        .withColumn("pc2", secondElement("pca"))
        .select(
            "appid", "name",
            "pc1", "pc2",
            "cluster",
            "pct_positive_reviews", "num_reviews", "price"
        )
        .toPandas()
)

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=plot_df,
    x="pc1",
    y="pc2",
    hue="cluster",
)
plt.title("K-Means Clusters of Games")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title="Cluster")
plt.show()
```

**Cluster 0 – Popular, Well-Rated Mainstream Titles**
Cluster 0 is the largest group (≈5,600 games) and contains many mainstream or broadly visible titles. ## **3b. Which words appear most frequently in positive and negative video-game reviews?**

```python
from pyspark.sql import functions as F
from pyspark.sql.functions import col, lower, regexp_replace, split, explode
from pyspark.ml.feature import StopWordsRemover
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = (
    reviews_and_games
    .select("review", "voted_up")
    .filter(col("review").isNotNull())
)

df_clean = (
    df
    .withColumn("review_clean",
                lower(regexp_replace(col("review"), "[^a-zA-Z\\s]", " ")))
    .withColumn("tokens", split(col("review_clean"), "\\s+"))
)

default_stops = StopWordsRemover.loadDefaultStopWords("english")
extra_stops = ["game", "games", "play", "played", "playing", "steam", ""]
all_stops = default_stops + extra_stops

remover = StopWordsRemover(
    inputCol="tokens",
    outputCol="filtered_tokens",
    stopWords=all_stops
)

df_tokens = remover.transform(df_clean)

words = (
    df_tokens
    .select("voted_up", explode(col("filtered_tokens")).alias("word"))
    .filter(col("word") != "")
)

word_sent = (
    words
    .groupBy("word", "voted_up")
    .count()
    .groupBy("word")
    .pivot("voted_up", [0, 1])   # 0 = negative, 1 = positive
    .agg(F.first("count"))
    .fillna(0, ["0", "1"])
    .withColumnRenamed("0", "neg_count")
    .withColumnRenamed("1", "pos_count")
)

totals = word_sent.agg(F.sum("pos_count"), F.sum("neg_count")).collect()[0]
total_pos = totals[0]
total_neg = totals[1]

alpha = 1.0

word_scored = (
    word_sent
    .withColumn("pos_rate", (col("pos_count") + alpha) / (total_pos + 2 * alpha))
    .withColumn("neg_rate", (col("neg_count") + alpha) / (total_neg + 2 * alpha))
    .withColumn("score", col("pos_rate") - col("neg_rate"))  # >0 = positive word
    .withColumn("total_count", col("pos_count") + col("neg_count"))
)

min_count = 3000
filtered_words = word_scored.filter(col("total_count") >= min_count)

positive_words = filtered_words.orderBy(col("score").desc()).limit(200)
negative_words = filtered_words.orderBy(col("score").asc()).limit(200)

pos_pd = positive_words.select("word", "pos_count").toPandas()
neg_pd = negative_words.select("word", "neg_count").toPandas()

pos_freq = dict(zip(pos_pd["word"], pos_pd["pos_count"]))
neg_freq = dict(zip(neg_pd["word"], neg_pd["neg_count"]))

pos_wc = WordCloud(
    width=1200, height=600, background_color="white"
).generate_from_frequencies(pos_freq)

neg_wc = WordCloud(
    width=1200, height=600, background_color="white"
).generate_from_frequencies(neg_freq)

plt.figure(figsize=(18, 9))
plt.imshow(pos_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Positive Sentiment Word Cloud")
plt.show()

plt.figure(figsize=(18, 9))
plt.imshow(neg_wc, interpolation="bilinear")
plt.axis("off")
plt.title("Negative Sentiment Word Cloud")
plt.show()
```

## **4c. Does lower playtime correlate with review sentiment?**

First, we want to bucket playtime:

```python
#from pyspark.sql import functions as F

reviews_playtime = (
    reviews_and_games
    .withColumn(
        "playtime_bucket",
        F.when(F.col("author_playtime_forever") < 60, "<1 hour")
         .when(F.col("author_playtime_forever") < 120, "1–2 hours")
         .when(F.col("author_playtime_forever") < 300, "2–5 hours")
         .when(F.col("author_playtime_forever") < 600, "5–10 hours")
         .otherwise("10+ hours")
    )
    .withColumn(
        "bucket_order",
        F.when(F.col("playtime_bucket") == "<1 hour", 1)
         .when(F.col("playtime_bucket") == "1–2 hours", 2)
         .when(F.col("playtime_bucket") == "2–5 hours", 3)
         .when(F.col("playtime_bucket") == "5–10 hours", 4)
         .otherwise(5)
    )
)
```

Next, we will count positive and negative reviews inside each bucket:

```python
import pandas as pd

sent_by_playtime = (
    reviews_playtime
        .groupBy("playtime_bucket", "bucket_order", "voted_up")
        .count()
        .orderBy("bucket_order")
)

sent_pd = sent_by_playtime.toPandas()
sent_pd["playtime_bucket"] = pd.Categorical(
    sent_pd["playtime_bucket"],
    categories=["<1 hour", "1–2 hours", "2–5 hours", "5–10 hours", "10+ hours"],
    ordered=True
)
sent_pd["sentiment"] = sent_pd["voted_up"].apply(lambda x: "Positive" if x == 1 else "Negative")
```

```python
#import seaborn as sns
#import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
sns.barplot(
    data=sent_pd,
    x="playtime_bucket",
    y="count",
    hue="sentiment",
    palette="pastel",
    errorbar=None
)
plt.title("Review Sentiment by Playtime Bucket")
plt.xlabel("Total Playtime")
plt.ylabel("Number of Reviews")
plt.xticks(rotation=25)
plt.show()
```

There are two main insights:
1. Players who invest a lot of time generally enjoy the game or feel satisfied enough to endorse it.
2. Short-playtime users are more likely to have negative sentiment because they bounced early or encountered friction.

We can see that the playtime–sentiment chart reveals a clear behavioral trend: long-playtime users overwhelmingly leave positive reviews,
while short-playtime users show more negativity, likely driven by refunds, early abandonment, or technical issues.

However, this pattern is influenced by survivorship bias i.e. only players who enjoy the game continue playing.

For this question, a hexbin chart was used instead of a scatterplot chart to avoid being dominated by zero-vote reviews.
